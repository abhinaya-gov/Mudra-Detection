```python
from ultralytics import YOLO
import cv2
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
print("="*60)
print("YOLO PHOTO HAND DETECTION")
print("="*60)

# Load your trained model
model_path = 'mudra_detection/weights/weights 13-12-2025 16_20_40_118/best.pt'
model = YOLO(model_path)

print(f"Loaded model: {model_path}\n")

# Input image path (change this to your image)
image_path = 'your_image'  # CHANGE THIS TO YOUR IMAGE PATH

# Check if image exists
if not Path(image_path).exists():
    print(f"Error: Image not found at {image_path}")
    print("Please provide a valid image path")
    exit()

print(f"ðŸ“· Processing image: {image_path}")

# Detection settings
conf_threshold = 0.5  # Confidence threshold (0-1)
save_output = True  # Save annotated image

# Run detection
results = model(image_path, conf=conf_threshold)

# Get results
result = results[0]
detections = result.boxes

# Print detection info
print(f"\n{'='*60}")
print("DETECTION RESULTS:")
print(f"{'='*60}")
print(f"Hands detected: {len(detections)}\n")

if len(detections) > 0:
    print(f"{'ID':<5} {'Confidence':<12} {'Box (x, y, w, h)'}")
    print("-" * 60)
    
    for i, box in enumerate(detections):
        conf = box.conf[0].item()
        xyxy = box.xyxy[0].cpu().numpy()
        x1, y1, x2, y2 = xyxy
        w, h = x2 - x1, y2 - y1
        
        print(f"{i+1:<5} {conf:.4f}       ({x1:.0f}, {y1:.0f}, {w:.0f}, {h:.0f})")
else:
    print("No hands detected")

# Save annotated image
if save_output:
    output_path = Path(image_path).stem + '_detected.jpg'
    annotated_img = result.plot()
    cv2.imwrite(output_path, annotated_img)
    print(f"\nâœ… Saved annotated image: {output_path}")

# Display image (optional - comment out if you don't want popup)
print("\nDisplaying image... (close window to continue)")
annotated_img_rgb = cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12, 8))
plt.imshow(annotated_img_rgb)
plt.axis('off')
plt.title(f'Hand Detection - {len(detections)} hand(s) detected')
plt.tight_layout()
plt.show()

print("\nâœ… Done!")
```
